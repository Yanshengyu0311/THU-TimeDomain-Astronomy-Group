```python
import numpy as np
import os
import matplotlib.pyplot as plt
spec_asii_file_name="dzsn2019np_-14.5.txt"

spec_data=np.loadtxt(spec_asii_file_name)
plt.plot(spec_data[:,0],spec_data[:,1])
```

```python
base_name=""
if spec_asii_file_name[-4:]=='.txt':
    base_name=spec_asii_file_name.strip('.txt')
if spec_asii_file_name[-4:]=='.dat':
    base_name=spec_asii_file_name.strip('.dat')
yaml_file=base_name+".yaml"
syn_file=base_name+"_syn.dat"
cmd="syn++ "+yaml_file+" > "+syn_file
os.system(cmd)
syn_data=np.loadtxt(syn_file)
offset=0.3
plt.plot(syn_data[:,0], syn_data[:,1]/syn_data[:,1].max())
plt.plot(spec_data[:,0],spec_data[:,1]/spec_data[:,1].max()+offset)
```
